```python
'''
class : `KNN`
init inputs : `X_train` (list[list[float]]), `y_train` (list[str])

description : This class stores the training data and classifies new data points using the KNN algorithm.
'''
class KNN:
    def __init__(self, X_train, y_train):
        self.features = np.array(X_train, dtype=np.float64)
        self.labels = y_train

    def train(self, X_train, y_train):
        self.features = X_train
        self.labels = y_train

    def euc_dist(self, point1, point2):
        return np.sqrt(np.sum((point1 - point2) ** 2))

    def classify(self, new_point, k=1):
        # combine features and labels into a single list
        combined = list(zip(self.features, self.labels))
        
        # Sort combined list by the Euclidean distance between new_point and each training point
        sorted_combined = sorted(combined, key=lambda x: self.euc_dist(x[0], new_point))
        
        # Get the k closest points
        k_closest = sorted_combined[:k]
        
        # Get the labels of the k closest points
        k_labels = [label for _, label in k_closest]
        
        # Count occurrences of each label
        from collections import Counter
        label_count = Counter(k_labels)
        
        # Find the majority class
        most_common = label_count.most_common()
        
        return most_common[0][0]


    def test(self, X_test, y_test, k=1):
        correct_classifications = 0
        total_classifications = len(X_test)
        
        for i in range(total_classifications):
            prediction = self.classify(X_test.iloc[i].to_numpy(), k)
            
            if prediction == y_test[i]:
                correct_classifications += 1
        
        # accuracy
        accuracy = correct_classifications / total_classifications
        return accuracy
```

```python
import pandas as pd
import numpy as np
from collections import defaultdict
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

df = pd.read_csv('/anvil/projects/tdm/data/iris/Iris.csv')
X = df.drop(['Species','Id'], axis=1)
y = df['Species']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

y_train = y_train.to_numpy()
y_test = y_test.to_numpy()
```

```python
class KNN:
    def __init__(self, X_train, y_train):
        self.features = X_train
        self.labels = y_train

    def train(self, X_train, y_train):
        self.features = X_train
        self.labels = y_train

    def euc_dist(self, point1, point2):
        point1 = np.array(point1)
        point2 = np.array(point2)
        return np.linalg.norm(point1 - point2)

    def classify(self, new_point, k=1):
        # sort the combined list by the distance from the point in the list to the new point
        nearest_labels = [x[1] for x in sorted(zip(self.features, self.labels), key=lambda x: self.euc_dist(x[0], new_point))[:k]]
        return max(set(nearest_labels), key=nearest_labels.count)

    def test(self, X_test, y_test, k=1):
        # short 1 line approach, efficent list comprehension
        return sum([1 for p in zip(X_test, y_test) if self.classify(p[0],k=k)==p[1]])/len(X_test)
        
    def scaled_distance(self, point1, point2, weights):
        # firstly, scale weights so they sum to 1 (each weight should be a fraction of the total 1)
        total_weight = sum(weights)
        normalized_weights = [w / total_weight for w in weights]
    
        # then, scale the 2 points by the weights (multiply each feature in the point by the corresponding weight)
        scaled_point1 = [p1 * w for p1, w in zip(point1, normalized_weights)]
        scaled_point2 = [p2 * w for p2, w in zip(point2, normalized_weights)]
    
        # finally, calculate and return the euclidean distance between the 2 points (use the existing euc_dist function)
        return self.euc_dist(scaled_point1, scaled_point2)

    def classify_weighted(self, new_point, k=1, weights=None):
        if weights is None:
            return self.classify(new_point, k=k)
    
        # now, write the classify function using the scaled_distance function
        nearest_labels = [x[1] for x in sorted(zip(self.features, self.labels), key=lambda x: self.scaled_distance(x[0], new_point, weights))[:k]]
        return max(set(nearest_labels), key=nearest_labels.count)
    
    def test_weighted(self, X_test, y_test, k=1, weights=None):
        return sum([1 for p in zip(X_test, y_test) if self.classify_weighted(p[0], k=k, weights=weights) == p[1]]) / len(X_test)

    def classify_distance(self, new_point, k=1, weights=None):
        # follow the same approach as the classify function. however, for each nearest neighbor, we need to save both the label and the distance
        # nearest_labels = [(label, distance), ... k times]
        if weights is None:
            distance_func = self.euc_dist
        else:
            distance_func = lambda x, y: self.scaled_distance(x, y, weights)
        
        distances = [(label, distance_func(feature, new_point)) for feature, label in zip(self.features, self.labels)]
        nearest_neighbors = sorted(distances, key=lambda x: x[1])[:k]

        # now, we need to select the class based on each distance, not just the label
        # we can find the average distance of each class and select the class with the smallest average distance

        class_distances = defaultdict(list)
    
        for label, dist in nearest_neighbors:
            class_distances[label].append(dist)
    
        class_avg_distance = {label: sum(dists) / len(dists) for label, dists in class_distances.items()}    

        return min(class_avg_distance, key=class_avg_distance.get)
```

1. It allows the model to learn from the training data by storing the feature values, X_train and their corresponding labels, y_train. The train function sets the features and labels attributes of the class.

2. When classifying a new point, KNN computes the the Euclidean distance between the new point and each point in the training dataset. Once they are are calculated, the algorithm sorts the points by distance in ascending order and selects the k closest points. The k neighbors' labels are then used to determine the predicted class for the new point.

3. In the event of a tie, the KNN algorithm selects a random class from the tied classes. In this code, the classify method uses the max(set(nearest_labels), key=nearest_labels.count) function, which counts the occurrences of each label in the nearest neighbors and picks the one that occurs the most. If there is a tie, max function returns the first one it encounters.

```python
def get_mse(y, y_pred):
    return np.mean((y - y_pred) ** 2)

def get_rmse(y, y_pred):
    return np.sqrt(get_mse(y, y_pred))

def get_mae(y, y_pred):
    return np.mean(np.abs(y - y_pred))

def get_r_squared(y, y_pred):
    ss_res = np.sum((y - y_pred) ** 2)
    ss_tot = np.sum((y - np.mean(y)) ** 2)
    return 1 - (ss_res / ss_tot)

mse = get_mse(y, y_pred)
rmse = get_rmse(y, y_pred)
mae = get_mae(y, y_pred)
r_squared = get_r_squared(y, y_pred)


print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
print(f'MAE: {mae}')
print(f'R-Squared: {r_squared}')
```

```python
def poly_predict(x, coeffs):
    # y_intercept is the last element in the array
    y_intercept = coeffs[-1]

    # predicted value can start as the y-intercept
    predicted_value = y_intercept

    # The rest of the elements are the coefficients, so we can determine the degree of the polynomial
    coeffs = coeffs[:-1]
    current_degree = len(coeffs)

    # Now, we can iterate through the coefficients and make a sum of coefficient * (x^current_degree)
    # remember that the first element in the array is the coefficient for the highest degree, and the last element is the coefficient for the lowest degree
    for i, coeff in enumerate(coeffs):
        predicted_value += coeff * (x ** current_degree)
        current_degree -= 1
        
    return predicted_value
```

The original 5th-degree polynomial had large coefficients, resulting in perfect accuracy but overfitting. LASSO shrinks some of the coefficients to zero as alpha increases, effectively performing feature selection. LASSO discards irrelevant polynomial terms by setting their coefficients to zero, leading to a simpler model that avoids overfitting.

Both LASSO and Ridge regression shrink coefficients as alpha increases, but they do it differently. LASSO sets some coefficients exactly to zero for feature selection, leading to sparser models. Ridge reduces the magnitude of all coefficients but does not set them to zero, meaning all polynomial terms remain in the model, though scaled down. As a result, LASSO tends to create simpler models by effectively removing less relevant polynomial terms, whereas Ridge keeps all terms but reduces their influence.

As the alpha value increases, the L1 value for both LASSO and Ridge decreases but LASSO's approach leads to a sparser model, while Ridge results in smoother but more complex models.

```python
class Perceptron:
    def __init__(self, learning_rate=0.01, n_epochs=1000, activation='relu'):
        self.learning_rate = learning_rate
        self.n_epochs = n_epochs
        self.activation = activation
        self.weights = None
        self.bias = None

    def activate(self, x):
        if self.activation == 'linear':
            return linear(x)
        elif self.activation == 'sigmoid':
            return sigmoid(x)
        elif self.activation == 'tanh':
            return tanh(x)
        else:  
            return relu(x)

    def activate_derivative(self, x):
        if self.activation == 'linear':
            return linear_d(x)
        elif self.activation == 'sigmoid':
            return sigmoid_d(x)
        elif self.activation == 'tanh':
            return tanh_d(x)
        else:  
            return relu_d(x)

    def compute_linear(self, X):
        return np.dot(X, self.weights) + self.bias

    def error(self, y_pred, y_true):
        return y_true - y_pred

    def backward_gradient(self, error, linear):
        gradient = -error * self.activate_derivative(linear)
        return gradient

    def predict(self, X):
        linear_output = self.compute_linear(X)
        return self.activate(linear_output)

    def train(self, X, y, reset_weights = True):
        if reset_weights or self.weights is None or self.bias is None:
            self.weights = np.zeros(X.shape[1])
            self.bias = 0

        for epoch in range(self.n_epochs):
            for i in range(X.shape[0]):
                linear_output = self.compute_linear(X[i])
                y_pred = self.activate(linear_output)
                error = self.error(y_pred, y[i])
                gradient = self.backward_gradient(error, linear_output)

                self.weights -= self.learning_rate * gradient * X[i]
                self.bias -= self.learning_rate * gradient


    def test(self, X, y):
        y_pred = self.predict(X)

        y = np.array(y).flatten()
        y_pred = np.array(y_pred).flatten()

        mean_squared_error = np.mean((y - y_pred) ** 2)
        root_mean_squared_error = np.sqrt(mean_squared_error)
        mean_absolute_error = np.mean(np.abs(y - y_pred))
    
        ss_res = np.sum((y - y_pred) ** 2)
        ss_tot = np.sum((y - np.mean(y)) ** 2)
        r_squared = 1 - (ss_res / ss_tot)
    

        return {
            "mean_squared_error": mean_squared_error,
            "root_mean_squared_error": root_mean_squared_error,
            "mean_absolute_error": mean_absolute_error,
            "r_squared": r_squared
        }
```
