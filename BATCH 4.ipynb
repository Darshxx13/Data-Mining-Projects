```python
import pandas as pd

def summarize_dataframe(dataframe):
    mean_dep_delay = dataframe['DepDelay'].mean()
    median_dep_delay = dataframe['DepDelay'].median()
    std_dev_dep_delay = dataframe['DepDelay'].std()
    shape = dataframe.shape
    
    print("MyDF Summary Statistics ---")
    print(f"Shape:({shape[0]} rows, {shape[1]} columns)")
    print(f"Mean: {mean_dep_delay:.3f}")
    print(f"Median: {median_dep_delay:.3f}")
    print(f"Standard Deviation: {std_dev_dep_delay:.3f}")
    print("---------------------------")

file_path = "/anvil/projects/tdm/data/flights/subset/1991.csv"
df = pd.read_csv(file_path)

summarize_dataframe(df)
```

```python
%%R

summarize_dataframe <- function(dataframe) {
    mean_dep_delay <- mean(dataframe$DepDelay, na.rm = TRUE)
    median_dep_delay <- median(dataframe$DepDelay, na.rm = TRUE)
    std_dev_dep_delay <- sd(dataframe$DepDelay, na.rm = TRUE)
    shape <- dim(dataframe)
    
    cat("MyDF Summary Statistics ---\n")
    cat("Shape:(", shape[1], "rows,", shape[2], "columns)\n")
    cat("Mean:", mean_dep_delay, "\n")
    cat("Median:", median_dep_delay, "\n")
    cat("Standard Deviation:", std_dev_dep_delay, "\n")
    cat("---------------------------\n")
}

file_path <- "/anvil/projects/tdm/data/flights/subset/1991.csv"
df <- read.csv(file_path)

summarize_dataframe(df)
```
